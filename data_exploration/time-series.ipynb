```python
import pandas as pd
import matplotlib.pyplot as plt

city = 'St. Louis'
review_data = pd.read_csv(f'../filtered_cities/{city}_review.csv')

review_data
```

```python
review_data['date'] = pd.to_datetime(review_data['date'])
review_data.set_index('date', inplace=True)
```

```python
# make a time series plot with the number of reviews per day
reviews_over_time = review_data.resample('D').size()

plt.figure(figsize=(10, 6))
reviews_over_time.plot(color='skyblue')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.show()
```

```python
# make a time series plot with the number of reviews per week
reviews_over_time = review_data.resample('W').size()

# print number of points on reviews_over_time
print(len(reviews_over_time))

plt.figure(figsize=(10, 6))
reviews_over_time.plot(color='skyblue')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.show()
```
