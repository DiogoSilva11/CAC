```python
import pandas as pd
from utils import load_filtered_data
```

```python
city = 'Jennings'
name = 'friends'

# load city data
data = load_filtered_data(city)
```

```python
# creating the nodes

nodes = data['user'].drop(columns=['friends'])
nodes = nodes.rename(columns={'user_id': 'id', 'name': 'people_name'})

nodes
```

```python
temp_edges_struct = {'from': [], 'to': [], 'weight': [], 'type': []}

for user_id, friends_str in data['user']['friends'].items():
    if isinstance(friends_str, str):
        friends_list = friends_str.split(', ')
        for friend_id in friends_list:
            temp_edges_struct['from'].append(user_id)
            temp_edges_struct['to'].append(friend_id)
            temp_edges_struct['weight'].append(1) 
            temp_edges_struct['type'].append('friendship')

edges = pd.DataFrame(temp_edges_struct)
edges = edges.drop_duplicates()

edges
```

```python
# save data
nodes.to_csv(f'nodes_and_edges/{city}_{name}_nodes.csv', index=False)
edges.to_csv(f'nodes_and_edges/{city}_{name}_edges.csv', index=False)
```
